```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year=[2014,2015,2016,2017,2018,2019,2020,2021] 
placement_Computer=[80,82,85,90,92,94,96,100] 
placement_Civil=[78,80,84,88,90,89,89,88]
```

```python
#plot line graph X axis = year; y- axis= placement statistics. Use the 
plt.figure(figsize=(6,4))
plt.plot(year, placement_Computer, label='Computer') 
plt.plot(year, placement_Civil, label='Civil')
plt.xlabel('Year')
plt.ylabel('Placement Statistics')
plt.title('Placement Statistics of Computer and Civil Engineering')
plt.legend() 
plt.show()
```

```python
#Scatter plot for random sine wave 
x=np.linspace(0,2*np.pi)
y=np.sin(x)
plt.figure(figsize=(5,3)) 
plt.scatter(x,y,label='Test')
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show()
```

```python
#plot barplot for same.
bar_width = 0.3
r1 = [i for i in range(len(year))] 
r2 = [i + bar_width for i in r1] 
plt.figure(figsize=(6,4))
plt.bar(r1, placement_Computer, color='green', width=bar_width,edgecolor='green')
plt.bar(r2, placement_Civil, color='red', width=bar_width,edgecolor='red')
plt.xlabel('Year')
plt.ylabel('Placement Statistics')
plt.title('Placement Statistics of Computer and Civil Engineering') 
plt.xticks([i + bar_width/2 for i in range(len(year))], year)
plt.show()
```

```python
#Histogram 
data=np.random.binomial(100,0.6,size=1000)
plt.figure(figsize=(5,3))
# Create a histogram with 20 bins 
plt.hist(data, bins=20)
# Set the x-axis label plt.xlabel('Xaxis Label')
# Set the y-axis label plt.ylabel('Yaxis Label')
# Set the title of the plot 
plt.title('Histogram')
# Display the plot 
plt.show()
```
