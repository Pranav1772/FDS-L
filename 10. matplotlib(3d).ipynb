```python
#create data for plot 
#Draw a 3d plot
import numpy as np
import matplotlib.pyplot as plt
z=np.linspace(0,np.pi*2)
x=np.sin(z) 
y=np.cos(z)
ax=plt.figure().add_subplot(projection='3d')
ax.plot(x,y,z) 
ax.set_xlabel('xaxis') 
ax.set_ylabel('y-axis') 
ax.set_zlabel('z-axis')
plt.show()
```

```python
# create data for the scatter plot 
x = np.random.normal(size=100) 
y = np.random.normal(size=100) 
z = np.random.normal(size=100) 
colors = np.random.rand(100)
# create a 3D figure 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# plot the scatter plot
ax.scatter(x, y, z, c=colors, cmap='viridis')
# set the labels for the axes 
ax.set_xlabel('X-axis') 
ax.set_ylabel('Yaxis') 
ax.set_zlabel('Z-axis')
```

```python

# x=np.linspace(0,2*np.pi)
# y=np.sin(x)
z= np.linspace(0,2*np.pi) 
x = np.cos(z-np.pi/2) 
y = np.sin(z-np.pi/2)
ax = plt.figure().add_subplot(projection='3d')
ax.stem(x, y, z, label='curve')
ax.legend() 
plt.show()
```

```python
# create data for the density and contour plot 
x = np.random.normal(size=1000) 
y = np.random.normal(size=1000) 
z = (x**2+y**2)/np.pi
# create a 3D figure 
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# plot the density and contour plot
hist, xedges, yedges = np.histogram2d(x, y, bins=30, density=True)
X, Y = np.meshgrid(xedges[:-1], yedges[:-1]) 
Z = hist.T
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8) 
ax.contour(X, Y, Z, offset=-0.15, cmap='viridis')
# set the labels for the axes 
ax.set_xlabel('X-axis') 
ax.set_ylabel('Y-axis') 
ax.set_zlabel('Z-axis')
#show plot 
plt.show()
```
